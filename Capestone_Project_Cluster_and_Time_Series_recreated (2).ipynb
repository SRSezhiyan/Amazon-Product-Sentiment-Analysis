```python
import pandas as pd
clean_data = pd.read_csv(r"/content/drive/MyDrive/Sports_and_outdoors_datafinal.csv")
clean_data
```

```python
clean_data.drop(columns=["Unnamed: 0"],inplace=True)
```

```python
clean_data.dropna(inplace=True)
```

```python
clean_data.rename(columns = {'Class':'Sentiment_Class'}, inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
pca = clean_data
pca = pca.astype(str)
le = LabelEncoder()
pca = pca.apply(le.fit_transform)
scaler = StandardScaler()
pca = scaler.fit_transform(pca)
```

```python
from sklearn.decomposition import PCA
dec = PCA(n_components=2)
dec.fit(pca)
X_transformed = dec.fit_transform(pca)
X_transformed
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss=[]
for k in range(2,12):
    kmeans=KMeans(n_clusters=k)
    kmeans.fit(X_transformed)
    wcss.append(kmeans.inertia_)
plt.plot(range(2,12),wcss)
plt.grid(True)
plt.xlabel('k')
plt.ylabel('wcss W.R.T inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
kmeans = KMeans(n_clusters=3,random_state=42)
labels = kmeans.fit_predict(X_transformed)
#silhouette_avg = silhouette_score(X_transformed,labels)
#print("For n_clusters =",3,"The average silhouette_score is :", silhouette_avg)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_transformed[labels == 0, 0], X_transformed[labels == 0, 1], s = 50, c = 'lightblue', label = 'Neutral Customers')
plt.scatter(X_transformed[labels == 1, 0], X_transformed[labels == 1, 1], s = 50, c = 'lightgreen', label = 'Positive Customers')
plt.scatter(X_transformed[labels == 2, 0], X_transformed[labels == 2, 1], s = 50, c = 'red', label = 'Negative Customers')
plt.title('Clusters of Customers')
plt.xlabel('customers')
plt.ylabel('Extracted data')
plt.legend()
plt.show()
```

```python
clean_data['reviewTime'] = pd.to_datetime(clean_data['reviewTime'])
time_series = clean_data[["overall","reviewTime"]]
time_series = time_series.set_index("reviewTime")
```

```python
df_monthly = time_series.resample('M').mean()
df_monthly.head(2)
df_monthly.dropna(inplace=True)
```

```python
time_series["overall"] = le.fit_transform(time_series["overall"])
```

```python
from statsmodels.tsa.stattools import adfuller,acf,pacf
def checkStationarity(data):
    pvalue = adfuller(data)[1]
    print(pvalue)
    if pvalue < 0.05:
        msg = "pvalue={}. Data is Stationary. Proceed to model building".format(pvalue)
    else:
        msg = "pvalue={}. Data is not Stationary. Make the data stationary before model building".format(pvalue)

    return msg
```

```python
print(checkStationarity(df_monthly))
```

```python
diff_data = df_monthly - df_monthly.shift()
diff_data
```

```python
diff_data.dropna(inplace=True)
print(checkStationarity(diff_data))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(df_monthly,period = 12)
d_trend= decomposition.trend
d_seasonal = decomposition.seasonal
d_residual = decomposition.resid
```

```python
fig,ax = plt.subplots(4,1,figsize=(20,12))

plt.subplot(411)
plt.plot(time_series,label = 'Temperature')
plt.legend()

plt.subplot(412)
plt.plot(d_trend,label = 'Trend')
plt.legend()

plt.subplot(413)
plt.plot(d_seasonal,label = 'Seasonal')
plt.legend()

plt.subplot(414)
plt.plot(d_residual,label = 'Residual')
plt.legend(loc = 'best')
```

```python
diff_data.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
plot_acf(df_monthly,lags=40);
plot_pacf(df_monthly,lags=40);
```

```python
plot_acf(diff_data,lags=40);
plot_pacf(diff_data,lags=40);
```

```python
import statsmodels.api as sm
for i in range(1,5):
    ar_m1 = sm.tsa.ARIMA(df_monthly,order = (i,0,0))
    r1 = ar_m1.fit()
    print(r1.summary())
```

```python
for i in range(1,5):
    ar_m1 = sm.tsa.ARIMA(df_monthly,order = (0,0,i))
    r1 = ar_m1.fit()
    print(r1.summary())
```

```python
p=4;q=2;d=1;
P=0;Q=1;D=1;
model2 = sm.tsa.statespace.SARIMAX(df_monthly,order=(p,d,q),seasonal_order=(P,D,Q,12)).fit()
model2.summary()
```

```python
pvalue = sm.stats.acorr_ljungbox(r1.resid,lags=[1],return_df=True)['lb_pvalue'].values
print (pvalue)
if pvalue < 0.05:
    print("Reject H0. Bad model")
else:
    print("Fail-to-Reject H0. Hence the selected model is Good model")
```

```python
t=int(len(df_monthly)*0.7)
train=df_monthly[:t]
test=df_monthly[t:]
```

```python
start = len(train)
end = len(train)+ len(test)-1

print('start = {},end = {}'.format(start,end))

predictions = model2.predict(start,end)
print(predictions)
```

```python
actual = []; pred = []
for i in range (len(predictions)):
    actual.append(test['overall'][i])
    pred.append(predictions[i])
    
df_res = pd.DataFrame({'actual': actual, 'predicted':pred})
df_res['err']=df_res.actual - df_res.predicted

print(df_res)
```

```python
final=sm.tsa.statespace.SARIMAX(test['overall'],order=(p,d,q),seasonal_order=(P,D,Q,12)).fit()
fcast = final.predict(start = len(test),end=len(test)+20)
print(fcast)
```

```python
test['overall'].plot(figsize=(10,8),legend=True,label = "Test Data")
train["overall"].plot(legend=True,label="Train Data")
predictions.plot(legend=True,label = "Predicted Data")
fcast.plot(legend=True,label ="Forecasted Data")
plt.show()
```
